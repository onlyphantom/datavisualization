```python
import pandas as pd
import altair as alt

alt.renderers.enable("mimetype")

df = pd.read_csv("https://raw.githubusercontent.com/onlyphantom/miband/main/data/run_1km.csv", parse_dates=['startTime'])
df['day_of_week'] = df['startTime'].dt.day_name()


alt.Chart(df).mark_point().encode(
    # x, y, color, shape, size
    # x='seconds_per_km:Q', quantitative (numerical data)
    # y='day_of_week:N', nominal (unordered, categorical)
    # O: ordinal types (rank-ordered)
    # T: temporal types (date/time)
    x='seconds_per_km',
    y='day_of_week'
).interactive()
```

```python
alt.Chart(df).mark_point().encode(
    # count, sum, min, max, average, median, stdev
    x='min(seconds_per_km)',
    y='day_of_week'
)
```

```python
alt.Chart(df).mark_point(color='#5f9ea0').encode(
    # count, sum, min, max, average, median, stdev
    alt.X('average(seconds_per_km)', title='Seconds/km (avg)', scale=alt.Scale(type='log')),
    alt.Y('day_of_week', axis=alt.Axis(title='day_of_week'))
)
```

```python
df['week'] = df['startTime'].dt.week
df.tail()
```

```python
base = alt.Chart(df).mark_line(color='lightpink').encode(
    alt.X('week', title='Week Number'),
    alt.Y('average(seconds_per_km)')
)


base + base.mark_circle(color='magenta')
```

```python
alt.Chart(df).mark_line(color='lightpink', point=True).encode(
    alt.X('week', title='Week Number'),
    alt.Y('average(seconds_per_km)')
)
```

```python
speed = alt.Chart(df).mark_line(color='#A9A9A9', point=True).encode(
    alt.X('week', title='Week Number'), 
    alt.Y('average(seconds_per_km)')
)

# dist = alt.Chart(df).mark_line().encode(
#     alt.X('week'),
#     alt.Y('sum(distance(m))')
# )

dist = speed.encode(
    alt.Y('sum(distance(m))')
)

# concatenation (| put them horizontally, & put them vertically)
dist | speed.interactive()
```

```python
# declarative: tell it 'what' you want, not 'how' you want it
speed = alt.Chart(df).mark_point(color='#A9A9A9').encode(
    alt.X('week', title='Week Number'), 
    alt.Y('average(seconds_per_km)'),
    tooltip=['date', 'distance(m)', 'maxPace(/meter)', 'seconds_per_km']
)

speed.interactive()
```

```python
alt.Chart(df).mark_bar().encode(
    alt.X('count(speed_per_km)'),
    alt.Y('hours(startTime):O')
)
```

```python
# create an interval selection over y-axis encoding
brush = alt.selection_interval(encodings=['y'])

# determine opacity based on brush
opacity = alt.condition(brush, alt.value(1), alt.value(0.2))

runs = alt.Chart(df).mark_bar(color='lightpink').encode(
    alt.X('count(speed_per_km)'),
    alt.Y('month(startTime):O', title=None, timeUnit='month'),
    tooltip=['month(startTime)', 'count(speed_per_km)'],
    opacity=opacity
).add_selection(
    brush
).properties(
    width=360,
    height=240
)

speed = alt.Chart(df).mark_circle(color='plum').encode(
    alt.X('week', title='Week Number'),
    alt.Y('seconds_per_km'),
    opacity=opacity
).properties(
    width=300,
    height=240
)


speed | runs
```
